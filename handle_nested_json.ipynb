```python
import pandas as pd
```

```python
dataset = [{ 
  "firstName": "Rack", 
  "lastName": "Jackon",
  "gender": "man",
  "age": 24,
  "address": 
    { 
       "streetAddress": "126",
       "city": "San Jone",
       "state": "CA",
       "postalCode": "394221"
    },
  "phoneNumbers": 
    [ 
      { 
         "type": "home",
         "number": "7383627627"
      } 
    ] 
},
{ 
  "firstName": "Ryan", 
  "lastName": "Seaol",
  "gender": "man",
  "age": 32,
  "address": 
    { 
       "streetAddress": "142",
       "city": "San Jone",
       "state": "WA",
       "postalCode": "394289"
    },
  "phoneNumbers": 
    [ 
      { 
         "type": "home",
         "number": "9998562698"
      } 
    ] 
}]
```

```python
df = pd.DataFrame(dataset)
df
```

```python
from pandas import json_normalize
```

```python
address_df_parsed = json_normalize(df['address'])
address_df_parsed
```

```python
type(address_df_parsed)
```

```python
parse_phn_df = pd.json_normalize(
    dataset,
    record_path =['phoneNumbers'],
    meta=['firstName', 'lastName', 'gender', 'age', 'address'],
    errors='ignore'
)
parse_phn_df
```

```python
street_df_parsed = json_normalize(dataset)
street_df_parsed
```

```python
merged_df = pd.merge(parse_phn_df,street_df_parsed, on=['firstName','lastName','age','gender'])
merged_df
```

```python
result_df = merged_df.drop(['address','phoneNumbers'], axis=1)
result_df
```
